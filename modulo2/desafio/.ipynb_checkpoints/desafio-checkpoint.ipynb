```python
import numpy as npm
import pandas as pd
```

```python
data = pd.read_csv("https://pycourse.s3.amazonaws.com/bike-sharing.csv")
```

#### Qual o tamanho do dataset?

```python
data.head()
```

```python
data.shape
```

#### Qual a média da coluna windspeed?

```python
data["windspeed"].mean()
```

#### Qual a média da coluna temp?

```python
data["temp"].mean()
```

#### Quantos registros existem para o ano de 2011?

```python
year = data["datetime"].apply(lambda d : d.split("-")[0])
year[year == "2011"].count()
```

#### Quantos registros existem para o ano de 2012?

```python
year = data["datetime"].apply(lambda d : d.split("-")[0])
year[year == "2012"].count()
```

#### Quantas locações de bicicletas foram efetuadas em 2011?

```python
data[data["year"] == 0]["total_count"].sum()
```

#### Quantas locações de bicicletas foram efetuadas em 2012?

```python
data[data["year"] == 1]["total_count"].sum()
```

#### Qual estação do ano contém a maior média de locações de bicicletas? E a menor?

```python
data.groupby(["season"]).mean()["total_count"]
```

#### Qual horário do dia contém a maior média de locações de bicicletas? E a menor?

```python
data.groupby(["hour"]).mean()["total_count"].sort_values()
```

#### Que dia da semana contém a maior média de locações de bicicletas? E a menor?

```python
data.groupby(["weekday"]).mean()["total_count"].sort_values()
```

#### Às quartas-feiras (weekday = 3), qual o horário do dia contém a maior média de locações de bicicletas? E a menor?

```python
data[data["weekday"] == 3].groupby(["hour"]).mean().sort_values(by="total_count")
```

#### Aos sábados (weekday = 6), qual o horário do dia contém a maior média de locações de bicicletas?

```python
data[data["weekday"] == 6].groupby(["hour"]).mean().sort_values(by="total_count")
```
